```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.patheffects as pe
from matplotlib.colors import ListedColormap, BoundaryNorm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score
from hmmlearn import hmm
import itertools
import re


# My settings
# my workdir
INPUT_DIR = r"c:\kazhydromet"      # read
OUTPUT_DIR = r"c:\kazhydromet\HMM" # write
os.makedirs(OUTPUT_DIR, exist_ok=True)
os.chdir(INPUT_DIR)

REGIONS = {
    'Аршалынский': 'Аршалы_Final.csv',
    'Ерейментауский': 'Ерейментау_Final.csv',
    'Жаксынский': 'Жаксы_Final.csv',
    'Коргалжынский': 'Коргалжын_Final.csv'
}

DISTRICT_NAME_MAP = {
    "Аршалынский": "Arshaly",
    "Ерейментауский": "Ereymentau",
    "Жаксынский": "Zhaksy",
    "Коргалжынский": "Korgalzhyn",
}

HMM_PARAMS = {
    'n_states': 3,
    'cov_type': 'diag',
    'n_iter': 200,
    'random_state': 42,
    'tol': 1e-3
}

ROBUSTNESS_PARAMS = {
    'n_runs': 50,
    'k_list': [2, 3, 4],
    'target': 'wheat',
    'window': 10
}

def save_to_output(filename):
    return os.path.join(OUTPUT_DIR, filename)


def load_season(region_name, climate_file, months=(4, 9), 
                yield_file='Урожайность.csv', sep=';', enc='utf-8'):
    urozh = pd.read_csv(yield_file, sep=sep, encoding=enc)
    urozh['yield_wheat'] = urozh['yield_wheat'].astype(str).str.replace(',', '.', regex=False).astype(float)
    y = urozh[urozh['Region'].astype(str).str.strip() == region_name.strip()].copy()

    clim = pd.read_csv(climate_file, sep=sep, header=None, encoding=enc)
    clim.columns = [f'col{i}' for i in range(clim.shape[1])]
    
    clim['Year'] = clim['col0'].astype(str).str.extract(r'(\d{4})', expand=False)
    clim = clim.dropna(subset=['Year'])
    clim['Year'] = clim['Year'].astype(int)
    
    clim['month'] = pd.to_numeric(clim['col15'], errors='coerce')
    clim['t_avg'] = pd.to_numeric(clim['col1'].astype(str).str.replace(',', '.', regex=False), errors='coerce')
    clim['precip'] = pd.to_numeric(clim['col14'].astype(str).str.replace(',', '.', regex=False), errors='coerce')
    clim['deficit'] = pd.to_numeric(clim['col9'].astype(str).str.replace(',', '.', regex=False), errors='coerce')

    m1, m2 = months
    season = (clim.loc[clim['month'].between(m1, m2, inclusive='both')]
              .groupby('Year', as_index=False)
              .agg(t_avg=('t_avg','mean'), precip=('precip','sum'), deficit=('deficit','mean')))

    data = (y.merge(season, on='Year', how='inner')
            .dropna(subset=['yield_wheat','t_avg','precip','deficit'])
            .sort_values('Year').reset_index(drop=True))
    
    return data



def brier_multiclass_from_transmat(A, states, average="mean"):
    A = np.asarray(A, dtype=float)
    states = np.asarray(states, dtype=int)
    K = A.shape[0]

    if A.shape != (K, K):
        raise ValueError(f"A must be square, got {A.shape}")
    if len(states) < 2:
        raise ValueError("Need >=2 states to score transitions.")

    p_next = A[states[:-1]]
    y_next = np.eye(K)[states[1:]]
    sq = (p_next - y_next) ** 2
    
    if average == "sum":
        return float(np.mean(np.sum(sq, axis=1)))
    elif average == "mean":
        return float(np.mean(np.mean(sq, axis=1)))
    else:
        raise ValueError("average must be 'sum' or 'mean'")

def brier_binary_lowstate(A, states, low_state):
    A = np.asarray(A, dtype=float)
    states = np.asarray(states, dtype=int)
    p_low = A[states[:-1], low_state]
    y_low = (states[1:] == low_state).astype(int)
    return float(np.mean((p_low - y_low) ** 2))

def infer_low_state_by_yield(data, state_col="state", yield_col="yield_wheat"):
    mean_y = data.groupby(state_col)[yield_col].mean()
    return int(mean_y.idxmin()), mean_y


def state_ranges_table(data, state_col="state"):
    g = data.groupby(state_col)
    states = sorted(data[state_col].unique())
    out = pd.DataFrame({state_col: states})
    out["n_years"] = [int(g.size().loc[s]) for s in states]

    def add_stats(col, prefix):
        out[f"{prefix}_mean"] = [g[col].mean().loc[s] for s in states]
        out[f"{prefix}_min"] = [g[col].min().loc[s] for s in states]
        out[f"{prefix}_q25"] = [g[col].quantile(0.25).loc[s] for s in states]
        out[f"{prefix}_med"] = [g[col].median().loc[s] for s in states]
        out[f"{prefix}_q75"] = [g[col].quantile(0.75).loc[s] for s in states]
        out[f"{prefix}_max"] = [g[col].max().loc[s] for s in states]

    add_stats("yield_wheat", "yield")
    add_stats("t_avg", "t")
    add_stats("precip", "precip")
    add_stats("deficit", "def")

    return out.round(2)

def fit_hmm_for_region(region_name, climate_file, save_ranges_csv=True, verbose=True, **hmm_params):
    data = load_season(region_name, climate_file)

    if verbose:
        print(f"\n {region_name.strip()} (N={len(data)} лет) ")
        print(data[['Year','yield_wheat','t_avg','precip','deficit']].round(2).to_string(index=False))

    X = data[['t_avg','precip','deficit']].values
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)

    model = hmm.GaussianHMM(
        n_components=hmm_params.get('n_states', 3),
        covariance_type=hmm_params.get('cov_type', 'diag'),
        n_iter=hmm_params.get('n_iter', 200),
        random_state=hmm_params.get('random_state', 42)
    )
    model.fit(Xs)

    states = model.predict(Xs)
    data["state"] = states
    A = model.transmat_

    safe_name = region_name.strip().replace(" ", "_")
    A_df = pd.DataFrame(A, index=[f"s{i}" for i in range(hmm_params.get('n_states', 3))],
                        columns=[f"s{i}" for i in range(hmm_params.get('n_states', 3))])
    A_path = save_to_output(f"{safe_name}_HMM_transmat.csv")
    A_df.to_csv(A_path, sep=";", encoding="utf-8-sig")
    if verbose:
        print(f"\nSaved: {A_path}")
    
    last_state = int(states[-1])
    p_next = A[last_state]
    low_state, mean_y = infer_low_state_by_yield(data, "state", "yield_wheat")
    risk_low_next = float(p_next[low_state])

    bs_sum = brier_multiclass_from_transmat(A, states, average="sum")
    bs_mean = brier_multiclass_from_transmat(A, states, average="mean")
    bs_low = brier_binary_lowstate(A, states, low_state)

    if verbose:
        print("\nTransition matrix A:")
        print(np.round(A, 3))
        print("\nAverage values by state (z-score):")
        print(np.round(model.means_, 2))
        print("\nStates by year:")
        print(data[['Year','yield_wheat','state']].to_string(index=False))
        print("\nP(state_of_the_next_year | state_of_the_last_year):")
        print(np.round(p_next, 3))
        print("\nBrier multiclass (sum):", round(bs_sum, 4))
        print("Brier multiclass (mean):", round(bs_mean, 4))
        print("\nAverage_yield_by_state:")
        print(mean_y.round(3).to_string())
        print("low_state =", low_state)
        print("Binary Brier (low_state):", round(bs_low, 4))

    ranges = state_ranges_table(data)
    if verbose:
        print("\n Ranges by status ")
        print(ranges.to_string(index=False))

    if save_ranges_csv:
        out_path = save_to_output(f"{safe_name}_HMM_state_ranges.csv")
        ranges.to_csv(out_path, sep=";", index=False, encoding="utf-8-sig")
        if verbose:
            print(f"\nSaved: {out_path}")

    metrics = {
        "district": region_name.strip(),
        "N_years": int(len(data)),
        "brier_mc_sum": float(bs_sum),
        "brier_mc_mean": float(bs_mean),
        "low_state": int(low_state),
        "risk_low_next": float(risk_low_next),
        "brier_low_binary": float(bs_low),
        "last_state": int(last_state),
        "p_next_s0": float(p_next[0]),
        "p_next_s1": float(p_next[1]),
        "p_next_s2": float(p_next[2]),
    }

    return data, model, scaler, ranges, metrics


def plot_risk_vs_brier(summary_df, x_col="brier_mc_mean", y_col="risk_low_next", 
                       name_col="district", out_png="HMM_Risk_vs_Brier.png"):
    if summary_df.empty:
        print("Nothing to plot: summary_df is empty.")
        return

    x = summary_df[x_col].to_numpy(dtype=float)
    y = summary_df[y_col].to_numpy(dtype=float)
    names_ru = summary_df[name_col].astype(str).to_list()
    names_en = [DISTRICT_NAME_MAP.get(n, n) for n in names_ru]
    median_brier = float(np.median(x))

    plt.figure(figsize=(8, 6), dpi=150)
    plt.scatter(x, y, s=80, edgecolors='black', zorder=3)

    for i, txt in enumerate(names_en):
        plt.annotate(txt, (x[i], y[i]), xytext=(5, 5), textcoords='offset points', fontsize=10)

    plt.axvline(median_brier, color='gray', linestyle='--', linewidth=1,
                label=f"Median Brier ({median_brier:.3f})")

    xpad, ypad = 0.01, 0.03
    plt.xlim(max(0.0, float(x.min()) - xpad), float(x.max()) + xpad)
    plt.ylim(max(0.0, float(y.min()) - ypad), min(1.0, float(y.max()) + ypad))

    plt.xlabel("Mean Brier score (lower is better)", fontsize=11)
    plt.ylabel("Probability of transition to low-yield regime", fontsize=11)
    plt.title("HMM transition risk vs. calibration by district", fontsize=12)
    plt.legend(loc='upper right', frameon=True)
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.tight_layout()
    plt.savefig(save_to_output(out_png), dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Saved plot: {save_to_output(out_png)}")

def build_median_matrix(all_results, prefix, region_order, state_labels=("C0", "C1", "C2")):
    regions = [r for r in region_order if r in all_results]
    if len(regions) == 0:
        regions = list(all_results.keys())

    states = list(state_labels)
    mat = np.full((len(regions), 3), np.nan, dtype=float)

    for i, reg in enumerate(regions):
        ranges = all_results[reg]["ranges"].sort_values("state").reset_index(drop=True)
        for s in [0, 1, 2]:
            row = ranges[ranges["state"] == s]
            if len(row) == 1:
                col = f"{prefix}_med"
                if col in row.columns:
                    mat[i, s] = float(row[col].iloc[0])

    return mat, regions, states

def plot_hmm_state_median_heatmaps_mono(all_results, region_order, cmap="Greys",
                                       save_path="FIG_HMM_state_medians_heatmap_MONO.png"):
    metrics = [
        ("yield", "Yield median (t/ha)"),
        ("t", "Temperature median (°C)"),
        ("precip", "Precipitation median (mm)"),
        ("def", "Moisture deficit median"),
    ]

    fig, axes = plt.subplots(2, 2, figsize=(12.5, 9), dpi=150)
    fig.suptitle("Median values of key variables by HMM state and district", fontsize=14)
    axes = axes.ravel()

    for ax, (prefix, title) in zip(axes, metrics):
        mat, regions_raw, states = build_median_matrix(all_results, prefix, region_order, 
                                                       state_labels=("C0", "C1", "C2"))
        regions = [DISTRICT_NAME_MAP.get(r, r) for r in regions_raw]

        im = ax.imshow(mat, aspect="auto", cmap=cmap)
        ax.set_title(title, fontsize=11)
        ax.set_xticks(np.arange(len(states)))
        ax.set_xticklabels(states)
        ax.set_yticks(np.arange(len(regions)))
        ax.set_yticklabels(regions)

        ax.set_xticks(np.arange(-0.5, len(states), 1), minor=True)
        ax.set_yticks(np.arange(-0.5, len(regions), 1), minor=True)
        ax.grid(which="minor", linestyle=":", linewidth=0.8)
        ax.tick_params(which="minor", bottom=False, left=False)

        for i in range(mat.shape[0]):
            for j in range(mat.shape[1]):
                if np.isfinite(mat[i, j]):
                    ax.text(j, i, f"{mat[i, j]:.2f}", ha="center", va="center", fontsize=9,
                           color="white", fontweight="medium",
                           path_effects=[pe.withStroke(linewidth=1.2, foreground="black")])

        cbar = plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
        cbar.ax.tick_params(labelsize=9)

    plt.tight_layout(rect=[0, 0, 1, 0.94])
    if save_path:
        plt.savefig(save_to_output(save_path), dpi=300, bbox_inches="tight")
        print("Saved:", save_to_output(save_path))
    plt.show()

def infer_state_to_dnf_from_ranges(ranges_df):
    def to_float(x):
        return pd.to_numeric(pd.Series(x).astype(str).str.replace(",", ".", regex=False), 
                           errors="coerce").iloc[0]
    
    r = ranges_df.copy()
    r.columns = [str(c).strip() for c in r.columns]

    if "state" not in r.columns:
        raise KeyError("ranges_df must contain a column 'state'.")

    if "def_med" in r.columns:
        r["score"] = r["def_med"].apply(to_float)
        ascending = False
    elif "yield_med" in r.columns:
        r["score"] = r["yield_med"].apply(to_float)
        ascending = True
    else:
        raise KeyError("Need 'def_med' or 'yield_med' in ranges.")

    r = r.dropna(subset=["score"]).sort_values("score", ascending=ascending).reset_index(drop=True)

    if r.shape[0] < 3:
        raise ValueError("3 states are expected in ranges.")

    dry_state = int(r.loc[0, "state"])
    fav_state = int(r.loc[2, "state"])
    mid_state = int(r.loc[1, "state"])

    return {dry_state: "D", mid_state: "N", fav_state: "F"}

def plot_regime_heatmap(all_results, district_order, 
                       out_path="HMM_regimes_heatmap_EN_VITERBI.png"):
    regime_rows = {}
    all_years = set()

    for reg_ru in district_order:
        pack = all_results[reg_ru]
        d = pack["data"][["Year", "state"]].copy()
        d["Year"] = pd.to_numeric(d["Year"], errors="coerce").astype(int)
        d["state"] = pd.to_numeric(d["state"], errors="coerce").astype(int)
        d = d.sort_values("Year").dropna().reset_index(drop=True)
        all_years.update(d["Year"].tolist())

        ranges = pack.get("ranges", None)
        if ranges is None:
            raise KeyError(f"{reg_ru}: нет pack['ranges']")

        mapping = infer_state_to_dnf_from_ranges(ranges)
        s = d.set_index("Year")["state"].map(mapping)
        regime_rows[reg_ru] = s

    years = sorted(all_years)
    df_regime = pd.DataFrame(regime_rows).T.reindex(index=district_order, columns=years).fillna("N")
    regions_en = [DISTRICT_NAME_MAP.get(r, r) for r in df_regime.index.tolist()]

    map_to_num = {"D": 0, "N": 1, "F": 2}
    Z = df_regime.replace(map_to_num).astype(int).values

    cmap = ListedColormap(["red", "yellow", "green"])
    norm = BoundaryNorm([-0.5, 0.5, 1.5, 2.5], cmap.N)

    fig, ax = plt.subplots(figsize=(14, 3.8), dpi=200)
    ax.imshow(Z, aspect="auto", cmap=cmap, norm=norm)

    ax.set_title("HMM Physical Regimes by Year (Viterbi-decoded)", pad=10)
    ax.set_xlabel("Year")
    ax.set_ylabel("Region")

    ax.set_xticks(np.arange(len(years)))
    ax.set_xticklabels(years, rotation=0)
    ax.set_yticks(np.arange(len(regions_en)))
    ax.set_yticklabels(regions_en)

    ax.set_xticks(np.arange(-.5, len(years), 1), minor=True)
    ax.set_yticks(np.arange(-.5, len(regions_en), 1), minor=True)
    ax.grid(which="minor", linestyle="-", linewidth=0.4)
    ax.tick_params(which="minor", bottom=False, left=False)

    legend_handles = [
        mpatches.Patch(color="red", label="D = Dry"),
        mpatches.Patch(color="yellow", label="N = Normal/Intermediate"),
        mpatches.Patch(color="green", label="F = Favorable"),
    ]
    ax.legend(handles=legend_handles, loc="upper center", bbox_to_anchor=(0.5, -0.22),
             ncol=3, frameon=False)

    plt.tight_layout()
    plt.savefig(save_to_output(out_path), dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", save_to_output(out_path))


def estimate_stickiness_increase(df, dry_state, window=10):
    df = df.sort_values("Year").reset_index(drop=True).copy()
    df["next_state"] = df["state"].shift(-1)

    df["stay_dry"] = ((df["state"] == dry_state) & (df["next_state"] == dry_state)).astype(float)
    df["in_dry"] = (df["state"] == dry_state).astype(float)

    roll_stay = df["stay_dry"].rolling(window, min_periods=window).sum()
    roll_in = df["in_dry"].rolling(window, min_periods=window).sum()

    a = np.where(roll_in.to_numpy() > 0, roll_stay.to_numpy() / roll_in.to_numpy(), np.nan)
    df["a_dry_dry_rolling"] = a

    s = pd.Series(df["a_dry_dry_rolling"]).dropna()
    if len(s) == 0:
        return np.nan, np.nan, np.nan, df

    base_a = float(s.iloc[0])
    last_a = float(s.iloc[-1])
    X_pct = np.nan if (base_a == 0 or np.isnan(base_a) or np.isnan(last_a)) else float((last_a - base_a) / base_a * 100)

    return base_a, last_a, X_pct, df

def analyze_dry_stickiness(all_results, window=10):
    stick_rows = []
    rolling_series = {}

    for reg, pack in all_results.items():
        d = pack["data"][["Year", "state"]].copy()
        dry_state = int(pack["metrics"]["low_state"])

        base_a, last_a, X_pct, d2 = estimate_stickiness_increase(d, dry_state=dry_state, window=window)

        stick_rows.append({
            "district": reg,
            "dry_state": dry_state,
            "a_dry_dry_base_window10": base_a,
            "a_dry_dry_last_window10": last_a,
            "X_pct_increase_stuck_dry_window10": X_pct,
            "n_years": int(d.shape[0]),
            "n_dry_years": int((d["state"] == dry_state).sum()),
        })

        rolling_series[reg] = d2
        safe = reg.strip().replace(" ", "_")
        d2.to_csv(save_to_output(f"{safe}_dry_stickiness_rolling.csv"), sep=";", index=False, encoding="utf-8-sig")

    stick_summary = pd.DataFrame(stick_rows).sort_values("district").round(6)
    
    print("\n DRY STICKINESS SUMMARY (rolling window = 10) ")
    print(stick_summary.to_string(index=False))
    
    stick_summary.to_csv(save_to_output("HMM_dry_stickiness_summary.csv"), sep=";", index=False, encoding="utf-8-sig")
    print("\nSaved: " + save_to_output("HMM_dry_stickiness_summary.csv"))

    return stick_summary, rolling_series

def plot_stickiness_rolling(rolling_series, district_order, 
                           out_png="HMM_dry_stickiness_rolling.png"):
    plt.figure(figsize=(10, 5), dpi=150)

    for reg in district_order:
        d2 = rolling_series[reg]
        plt.plot(d2["Year"], d2["a_dry_dry_rolling"], marker="o", linewidth=1, label=DISTRICT_NAME_MAP.get(reg, reg))

    plt.ylim(-0.05, 1.05)
    plt.xlabel("Year")
    plt.ylabel("Rolling a_dry,dry (window=10)")
    plt.title("Probability of staying in low-yield regime (rolling estimate)")
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.savefig(save_to_output(out_png), dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved plot:", save_to_output(out_png))

def plot_risk_brier_stickiness(summary, stick_summary, district_order,
                               out_path="HMM_risk_brier_stickiness.png"):
    def to_float_series(s):
        return pd.to_numeric(s.astype(str).str.replace(",", ".", regex=False), errors="coerce")

    summ = summary.copy()
    stick = stick_summary.copy()

    for col in ["risk_low_next", "brier_low_binary"]:
        if col in summ.columns:
            summ[col] = to_float_series(summ[col])

    for col in ["a_dry_dry_last_window10"]:
        if col in stick.columns:
            stick[col] = to_float_series(stick[col])

    df = (summ[["district", "risk_low_next", "brier_low_binary"]]
          .merge(stick[["district", "a_dry_dry_last_window10"]], on="district", how="left")
          .rename(columns={
              "risk_low_next": "RiskDryNext",
              "brier_low_binary": "BrierLow",
              "a_dry_dry_last_window10": "Stickiness"
          }))

    df["district"] = df["district"].astype(str).str.strip()
    df = df.set_index("district").reindex(district_order).reset_index()
    df["District"] = df["district"].map(lambda x: DISTRICT_NAME_MAP.get(x, x))
    df["Risk_pct"] = df["RiskDryNext"] * 100.0

    x = np.arange(len(df))
    fig, ax1 = plt.subplots(figsize=(10, 4.8), dpi=200)

    bars = ax1.bar(x, df["Risk_pct"])
    ax1.set_xticks(x)
    ax1.set_xticklabels(df["District"], rotation=0)
    ax1.set_xlabel("District")
    ax1.set_ylabel("Risk of low-yield regime next year (%)")
    ax1.set_title("HMM risk (bars) with Brier(Low) and Dry stickiness (markers)")

    for rect, val in zip(bars, df["Risk_pct"]):
        if np.isfinite(val):
            ax1.text(rect.get_x() + rect.get_width() / 2, rect.get_height() + 1.0,
                    f"{val:.1f}%", ha="center", va="bottom", fontsize=9)

    ax2 = ax1.twinx()

    line_brier, = ax2.plot(x, df["BrierLow"], marker="o", linewidth=1.8,
                           color="green", label="Brier score (Low event)")
    line_stick, = ax2.plot(x, df["Stickiness"], marker="s", linewidth=1.8,
                          color="red", label="Dry stickiness (last 10-year window)")

    ax2.set_ylabel("Score / probability (0..1)")
    ax2.set_ylim(-0.05, 1.05)

    for xi, y in zip(x, df["BrierLow"]):
        if np.isfinite(y):
            ax2.text(xi, y + 0.03, f"{y:.3f}", ha="center", va="bottom", fontsize=9, color="green")
    for xi, y in zip(x, df["Stickiness"]):
        if np.isfinite(y):
            ax2.text(xi, y - 0.05, f"{y:.3f}", ha="center", va="top", fontsize=9, color="red")

    ax2.legend([line_brier, line_stick],
              ["Brier score (Low event)", "Dry stickiness (last 10-year window)"],
              loc="upper center", bbox_to_anchor=(0.5, -0.18), ncol=2, frameon=False)

    plt.tight_layout()
    plt.savefig(save_to_output(out_path), dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", save_to_output(out_path))


def build_district_dfs_from_all_results(all_results, target="wheat"):
    def pick_first(cols, candidates):
        for c in candidates:
            if c in cols:
                return c
        return None

    def pick_by_regex(cols, patterns):
        for pat in patterns:
            rx = re.compile(pat, flags=re.IGNORECASE)
            for c in cols:
                if rx.search(str(c)):
                    return c
        return None

    district_dfs = {}
    report = []

    for district, pack in all_results.items():
        if "data" not in pack:
            report.append((district, "SKIP: no 'data' in pack", None))
            continue

        df = pack["data"].copy()
        cols = df.columns.tolist()

        year_col = pick_first(cols, ["Year", "year", "YEAR"]) or pick_by_regex(cols, [r"\byear\b"])

        if target == "wheat":
            yield_candidates = ["yield_wheat", "Yield_wheat", "wheat_yield", "yield"]
        else:
            yield_candidates = ["yield_cereals", "Yield_cereals", "cereals_yield", "yield"]

        yield_col = pick_first(cols, yield_candidates) or pick_by_regex(cols, [r"yield"])

        temp_col = pick_first(cols, ["t_avg", "t_mean", "temp", "Temp"]) or pick_by_regex(
            cols, [r"\bt[_\s]*avg\b", r"\bt[_\s]*mean\b", r"\btemp\b"])

        precip_col = pick_first(cols, ["precip", "ppt", "prcp", "precip_sum"]) or pick_by_regex(
            cols, [r"\bprecip\b", r"\bppt\b", r"\bprcp\b", r"precip[_\s]*sum"])

        deficit_col = pick_first(cols, ["deficit", "def", "water_deficit", "moisture_deficit"]) or pick_by_regex(
            cols, [r"\bdeficit\b", r"\bdef\b"])

        missing = [name for name, col in [
            ("Year", year_col), ("Yield", yield_col), ("Temp", temp_col),
            ("Precip", precip_col), ("Deficit", deficit_col),
        ] if col is None]

        if missing:
            report.append((district, f"MISSING: {missing}", cols))
            continue

        use_cols = [year_col, yield_col, temp_col, precip_col, deficit_col]
        out = (df[use_cols].dropna().sort_values(year_col).reset_index(drop=True).copy())

        out = out.rename(columns={
            year_col: "Year", yield_col: "yield", temp_col: "temp",
            precip_col: "precip", deficit_col: "deficit"
        })

        district_dfs[district] = out
        report.append((district, "OK", out.columns.tolist()))

    return district_dfs, report

def fit_hmm_states(df, feature_cols, k=3, seed=42, scale=True, cov_type="diag", n_iter=200, tol=1e-3):
    df = df.sort_values("Year").reset_index(drop=True).copy()
    X = df[feature_cols].astype(float).values

    scaler = None
    if scale:
        scaler = StandardScaler()
        X = scaler.fit_transform(X)

    model = hmm.GaussianHMM(n_components=k, covariance_type=cov_type, 
                           n_iter=n_iter, tol=tol, random_state=seed)
    model.fit(X)
    states = model.predict(X)
    years = df["Year"].values
    return years, states, model, scaler

def best_perm_ari(ref_states, cand_states, k):
    best_ari = -1.0
    best_perm = None
    for perm in itertools.permutations(range(k)):
        mapping = {old: new for old, new in enumerate(perm)}
        mapped = np.vectorize(mapping.get)(cand_states)
        ari = adjusted_rand_score(ref_states, mapped)
        if ari > best_ari:
            best_ari = ari
            best_perm = perm
    return best_ari, best_perm

def approx_n_params_gaussian_hmm(k, d, cov_type="diag"):
    p_start = (k - 1)
    p_trans = k * (k - 1)
    p_means = k * d
    if cov_type == "full":
        p_cov = k * (d * (d + 1) / 2)
    else:
        p_cov = k * d
    return int(p_start + p_trans + p_means + p_cov)

def aic_bic(logL, n_params, n_obs):
    if np.isnan(logL):
        return np.nan, np.nan
    aic = 2*n_params - 2*logL
    bic = n_params*np.log(n_obs) - 2*logL
    return aic, bic

def run_robustness_checks(all_results, feature_cols=["temp", "precip", "deficit"], 
                         n_runs=50, k_list=[2, 3, 4], base_k=3, scale=True,
                         cov_type="diag", n_iter=200, tol=1e-3):
    district_dfs, report = build_district_dfs_from_all_results(all_results, target="wheat")
    
    print(" Column mapping / load report ")
    for d, status, info in report:
        if status == "OK":
            print(f" - {d}: OK -> {info}")
        else:
            print(f" - {d}: {status}")

    if len(district_dfs) == 0:
        print("\nSTOP: district_dfs is empty")
        return None

    seeds = list(range(n_runs))

    init_rows = []
    for district, df in district_dfs.items():
        _, states_ref, _, _ = fit_hmm_states(df, feature_cols, k=base_k, seed=42, scale=scale,
                                            cov_type=cov_type, n_iter=n_iter, tol=tol)
        for seed in seeds:
            _, states_c, _, _ = fit_hmm_states(df, feature_cols, k=base_k, seed=seed, scale=scale,
                                              cov_type=cov_type, n_iter=n_iter, tol=tol)
            ari, perm = best_perm_ari(states_ref, states_c, k=base_k)
            init_rows.append({"district": district, "seed": seed, "ARI_vs_baseline": ari, "best_perm": str(perm)})

    init_df = pd.DataFrame(init_rows)
    init_summary = (init_df.groupby("district")["ARI_vs_baseline"]
                   .agg(median="median", q25=lambda s: s.quantile(0.25), 
                        q75=lambda s: s.quantile(0.75), min="min", max="max", 
                        mean="mean", std="std").reset_index())

    print("\n Robustness 1) Sensitivity to initialization (K=3): ARI vs baseline ")
    print(init_summary.to_string(index=False))

    loyo_rows = []
    for district, df in district_dfs.items():
        df = df.sort_values("Year").reset_index(drop=True).copy()
        years_ref, states_ref, _, _ = fit_hmm_states(df, feature_cols, k=base_k, seed=42, scale=scale,
                                                     cov_type=cov_type, n_iter=n_iter, tol=tol)
        years_all = df["Year"].tolist()

        for i in range(len(df)):
            left_out_year = int(df.loc[i, "Year"])
            df_train = df.drop(index=i).reset_index(drop=True)
            years_loyo, states_loyo, _, _ = fit_hmm_states(df_train, feature_cols, k=base_k, seed=42, 
                                                           scale=scale, cov_type=cov_type, 
                                                           n_iter=n_iter, tol=tol)

            train_years = df_train["Year"].tolist()
            idxs = [years_all.index(y) for y in train_years]
            states_ref_train = states_ref[idxs]

            ari, perm = best_perm_ari(states_ref_train, states_loyo, k=base_k)
            loyo_rows.append({"district": district, "left_out_year": left_out_year, 
                            "ARI_LOYO_vs_baseline": ari, "best_perm": str(perm)})

    loyo_df = pd.DataFrame(loyo_rows)
    loyo_summary = (loyo_df.groupby("district")["ARI_LOYO_vs_baseline"]
                   .agg(median="median", q25=lambda s: s.quantile(0.25), 
                        q75=lambda s: s.quantile(0.75), min="min", max="max", 
                        mean="mean", std="std").reset_index())

    print("\n Robustness 2) LOYO (K=3): ARI vs baseline ")
    print(loyo_summary.to_string(index=False))

    ablation_rows = []
    for district, df in district_dfs.items():
        _, states_ref, _, _ = fit_hmm_states(df, feature_cols, k=base_k, seed=42, scale=scale,
                                            cov_type=cov_type, n_iter=n_iter, tol=tol)
        for drop_f in feature_cols:
            cols = [c for c in feature_cols if c != drop_f]
            _, states_c, _, _ = fit_hmm_states(df, cols, k=base_k, seed=42, scale=scale,
                                              cov_type=cov_type, n_iter=n_iter, tol=tol)
            ari, perm = best_perm_ari(states_ref, states_c, k=base_k)
            ablation_rows.append({"district": district, "dropped_feature": drop_f, 
                                "ARI_vs_baseline": ari, "best_perm": str(perm)})

    ablation_df = pd.DataFrame(ablation_rows)
    print("\n Robustness 3) Feature ablation (drop-one): ARI vs baseline ")
    print(ablation_df.to_string(index=False))

    k_rows = []
    for district, df in district_dfs.items():
        df = df.sort_values("Year").reset_index(drop=True).copy()
        X_raw = df[feature_cols].astype(float).values
        X_eval = StandardScaler().fit_transform(X_raw) if scale else X_raw
        d = len(feature_cols)

        for k in k_list:
            _, _, model_k, _ = fit_hmm_states(df, feature_cols, k=k, seed=42, scale=scale,
                                             cov_type=cov_type, n_iter=n_iter, tol=tol)
            try:
                logL = model_k.score(X_eval)
            except Exception:
                logL = np.nan

            p = approx_n_params_gaussian_hmm(k, d, cov_type=cov_type)
            aic, bic = aic_bic(logL, p, n_obs=len(df))

            k_rows.append({"district": district, "K": k, "logL": logL, 
                         "AIC": aic, "BIC": bic, "n_params_approx": p})

    k_df = pd.DataFrame(k_rows)
    print("\n Robustness 4) K variation (2..4): logL/AIC/BIC ")
    print(k_df.to_string(index=False))
    
    init_summary.to_csv(save_to_output("HMM_robustness_init_ARI_summary.csv"), index=False, encoding="utf-8-sig")
    loyo_summary.to_csv(save_to_output("HMM_robustness_LOYO_ARI_summary.csv"), index=False, encoding="utf-8-sig")
    ablation_df.to_csv(save_to_output("HMM_robustness_ablation_ARI.csv"), index=False, encoding="utf-8-sig")
    k_df.to_csv(save_to_output("HMM_robustness_K_variation.csv"), index=False, encoding="utf-8-sig")

    print("\nSaved CSV:")
    print(" - " + save_to_output("HMM_robustness_init_ARI_summary.csv"))
    print(" - " + save_to_output("HMM_robustness_LOYO_ARI_summary.csv"))
    print(" - " + save_to_output("HMM_robustness_ablation_ARI.csv"))
    print(" - " + save_to_output("HMM_robustness_K_variation.csv"))

    return init_summary, loyo_summary, ablation_df, k_df

def plot_robustness_results():
    def to_float_series(s):
        return pd.to_numeric(s.astype(str).str.replace(",", ".", regex=False), errors="coerce")

    def norm_colname(c: str) -> str:
        return str(c).strip().lower().replace("\ufeff", "")

    def find_col(df, candidates):
        cols = {norm_colname(c): c for c in df.columns}
        for cand in candidates:
            cand_n = norm_colname(cand)
            if cand_n in cols:
                return cols[cand_n]
        for c_norm, c_orig in cols.items():
            for cand in candidates:
                if norm_colname(cand) in c_norm:
                    return c_orig
        raise KeyError(f"Не нашёл столбец. Ищу: {candidates}. Есть: {list(df.columns)}")

    init = pd.read_csv(save_to_output("HMM_robustness_init_ARI_summary.csv"), encoding="utf-8-sig")
    loyo = pd.read_csv(save_to_output("HMM_robustness_LOYO_ARI_summary.csv"), encoding="utf-8-sig")
    abl = pd.read_csv(save_to_output("HMM_robustness_ablation_ARI.csv"), encoding="utf-8-sig")

    init_district_col = find_col(init, ["района", "район", "district"])
    loyo_district_col = find_col(loyo, ["районная", "района", "район", "district"])
    init_med_col = find_col(init, ["медиана", "медиана района", "районная медиана", "median"])
    loyo_med_col = find_col(loyo, ["медиана", "районная медиана", "медиана района", "median"])

    init_med = to_float_series(init[init_med_col])
    loyo_med = to_float_series(loyo[loyo_med_col])

    district_order = list(REGIONS.keys())

    init_s = pd.DataFrame({"district": init[init_district_col], "med": init_med}).set_index("district").reindex(district_order)
    loyo_s = pd.DataFrame({"district": loyo[loyo_district_col], "med": loyo_med}).set_index("district").reindex(district_order)

    hmm_vals = init_s["med"].to_numpy()
    loyo_vals = loyo_s["med"].to_numpy()

    abl = abl.rename(columns={c: c.strip() for c in abl.columns})
    district_col = find_col(abl, ["район", "district"])
    feature_col = find_col(abl, ["с пониженной характеристикой", "dropped_feature", "feature"])
    ari_col = find_col(abl, ["ari_vs_baseline", "ARI_vs_baseline"])

    abl["district"] = abl[district_col].astype(str).str.strip()
    abl["feature_raw"] = abl[feature_col].astype(str).str.strip().str.lower()
    abl["ari"] = to_float_series(abl[ari_col])

    def norm_feature(x: str) -> str:
        if ("temp" in x) or ("темп" in x) or ("температ" in x):
            return "temperature"
        if ("precip" in x) or ("осад" in x):
            return "precipitation"
        if ("deficit" in x) or ("дефиц" in x):
            return "moisture_deficit"
        return "other"

    abl["feature"] = abl["feature_raw"].map(norm_feature)
    abl = abl[abl["feature"] != "other"].copy()

    pivot = (abl.pivot_table(index="district", columns="feature", values="ari", aggfunc="first")
            .reindex(district_order))

    temp_vals = pivot["temperature"].to_numpy() if "temperature" in pivot.columns else np.full(len(pivot), np.nan)
    prec = pivot["precipitation"].to_numpy() if "precipitation" in pivot.columns else np.full(len(pivot), np.nan)
    defi = pivot["moisture_deficit"].to_numpy() if "moisture_deficit" in pivot.columns else np.full(len(pivot), np.nan)

    pm_vals = np.nanmin(np.vstack([prec, defi]), axis=0)
    only_one = np.isnan(pm_vals)
    pm_vals[only_one] = np.nanmax(np.vstack([prec, defi]), axis=0)[only_one]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    plt.subplots_adjust(wspace=0.25)

    x = np.arange(len(district_order))
    w = 0.35

    b1 = ax1.bar(x - w/2, hmm_vals, width=w, label="HMM ARI")
    b2 = ax1.bar(x + w/2, loyo_vals, width=w, label="LOYO ARI")

    ax1.set_title("Model Stability by District")
    ax1.set_xlabel("District")
    ax1.set_ylabel("ARI")
    ax1.set_xticks(x)
    ax1.set_xticklabels([DISTRICT_NAME_MAP[d] for d in district_order])
    ax1.set_ylim(0, 1.0)
    ax1.legend(title="Measure", loc="upper right")

    for bars in (b1, b2):
        for bar in bars:
            y = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2, y + 0.02, f"{y:.2f}",
                    ha="center", va="bottom", fontsize=9)

    x2 = np.arange(len(district_order))
    w2 = 0.35

    c1 = ax2.bar(x2 - w2/2, temp_vals, width=w2, label="Temperature")
    c2 = ax2.bar(x2 + w2/2, pm_vals, width=w2, label="Precipitation/Moisture")

    ax2.set_title("Impact of Excluding Features on ARI")
    ax2.set_xlabel("District")
    ax2.set_ylabel("ARI")
    ax2.set_xticks(x2)
    ax2.set_xticklabels([DISTRICT_NAME_MAP[d] for d in district_order])
    ax2.set_ylim(0, 1.0)
    ax2.legend(title="Feature", loc="upper right")

    for bars in (c1, c2):
        for bar in bars:
            y = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2, y + 0.02, f"{y:.2f}",
                    ha="center", va="bottom", fontsize=9)

    plt.savefig(save_to_output("HMM_robustness_comparison.png"), dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", save_to_output("HMM_robustness_comparison.png"))


def main():
    print("HMM Analysis")
    print(f"Input data: {INPUT_DIR}")
    print(f"Output data: {OUTPUT_DIR}")

    all_results = {}
    metrics_rows = []

    for reg, f in REGIONS.items():
        try:
            data, model, scaler, ranges, metrics = fit_hmm_for_region(
                reg, f, verbose=True, **HMM_PARAMS
            )
            all_results[reg] = {
                "data": data, "model": model, "scaler": scaler,
                "ranges": ranges, "metrics": metrics
            }
            metrics_rows.append(metrics)
        except Exception as e:
            print(f"\n❌ Ошибка {reg}: {e}")

    if not metrics_rows:
        print("No metrics computed; check input files and region names.")
        return

    summary = pd.DataFrame(metrics_rows).sort_values("district").round(6)
    print("\n\n Summary (HMM + Brier + Risk) ")
    print(summary[["district","N_years","low_state","risk_low_next",
                   "brier_mc_mean","brier_low_binary","last_state"]]
          .round(4).to_string(index=False))

    summary.to_csv(save_to_output("HMM_BRIER_summary.csv"), sep=";", index=False, encoding="utf-8-sig")
    print("\nSaved: " + save_to_output("HMM_BRIER_summary.csv"))

    
    plot_risk_vs_brier(summary)

    
    plot_hmm_state_median_heatmaps_mono(
        all_results,
        region_order=list(REGIONS.keys()),
        cmap="Greys"
    )

    
    plot_regime_heatmap(all_results, list(REGIONS.keys()))

    
    stick_summary, rolling_series = analyze_dry_stickiness(
        all_results, 
        window=ROBUSTNESS_PARAMS['window']
    )
    
    plot_stickiness_rolling(rolling_series, list(REGIONS.keys()))
    
    plot_risk_brier_stickiness(summary, stick_summary, list(REGIONS.keys()))

    
    print("\n" )
    print("ROBUSTNESS CHECKS")
    
    run_robustness_checks(
        all_results,
        feature_cols=["temp", "precip", "deficit"],
        n_runs=ROBUSTNESS_PARAMS['n_runs'],
        k_list=ROBUSTNESS_PARAMS['k_list'],
        base_k=HMM_PARAMS['n_states'],
        scale=True,
        cov_type=HMM_PARAMS['cov_type'],
        n_iter=HMM_PARAMS['n_iter'],
        tol=HMM_PARAMS['tol']
    )
    
    plot_robustness_results()

    print("\n")
    print("Finish!")
    print("All results saved to: " + OUTPUT_DIR)

if __name__ == "__main__":
    main()
```
